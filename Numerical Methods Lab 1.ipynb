2c.

```python
f2(x)=diff(f(x),x,2)
print("f2(x)=",f2(x))
```

```python
f3(x)=diff(f(x),3)
f4(x)=diff(f(x),4)
```

```python
fext1 = (solve(f4(x)==0,x))
print(fext1)
print("\n")
fext2 = f3(-1.5)
fext3 = f3(1.5)
fext4 = f3(-math.sqrt(1/2*math.sqrt(6) + 3/2))
fext5 = f3(math.sqrt(1/2*math.sqrt(6) + 3/2))
fext6 = f3(-math.sqrt(-1/2*math.sqrt(6) + 3/2))
fext7 = f3(math.sqrt(-1/2*math.sqrt(6) + 3/2))

fmax1 = max(fext2,fext3,fext4,fext5,fext6,fext7)

feb1 = abs((fmax1)/(factorial(3))*(1.5)^3)
print("Error bound =",feb1)
```

```python
g(x)=sin(x)
T3(x)=taylor(g(x),x,math.pi/4,3)
print("T3(x) =",T3(x))
```

```python
g4(x)=diff(g(x),4)
g5(X)=diff(g(x),5)
```

```python
gext1 = (solve(g5(x)==0,x))
gext2 = g4(-math.pi/4)
gext3 = g4(3*math.pi/4)
gext4 = g4(math.pi/2)

print(gext1)
print("\n")
gext5 = max(gext2,gext3,gext4)

geb1 = abs((gext5)/(factorial(4))*(3*math.pi/4-math.pi/4)^4)
print("Error Bound =",geb1)
```
